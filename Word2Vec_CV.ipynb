```python
# Load corpus
import pandas as pd

data = pd.read_csv("webtext.csv")
l_text = list(data.text)
```

```python
def t_r(text):
    text = re.sub('<body>','',text)
    text = re.sub('</body>','',text)
    text = re.sub('\W+',' ',text)
    text = re.sub('(\W+)(\d+)(\W+)',' ',text)
    text = text.split()

    return text
```

```python
import re

sen=list()
for i in l_text:
    data1= t_r(str(i))
    sen.append(data1)
print(sen)
```

```python
def _len(item):
    if(15>len(item)>2):
        return True
    else:
        return False
    
fil_sen = list(filter(_len, sen))
```

```python
from gensim.models import Word2Vec

sample = Word2Vec(fil_sen, min_count=1, size=5)

l = ['nice', 'good', 'book', 'read']
rep = [sample.wv.get_vector(i) for i in l]
for word, rep in zip(l, rep):
    print(word, rep)
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer

corpus = []
for i in range(len(fil_sen)):
    ps = PorterStemmer()
    review = [ps.stem(word) for word in fil_sen[i] if not word in set(stopwords.words('english'))]
    sens = ' '.join(fil_sen[i])
    corpus.append(sens)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 200)
X = cv.fit_transform(corpus).toarray()
```
